Ukuran sebaran (*measure of dispersion*) adalah statistika deskriptif yang digunakan untuk membantu kita memahami sebaran titik data di dalam sebuah sampel ataupun populasi. Terdapat beberapa ukuran sebaran yang biasanya digunakan tergantung pada jenis atau tipe datanya, yaitu: 1. Tipe Data Nominal dan Ordinal; A. Proporsi Kategori (*Categorical Proportion*), B. Persen Proporsi (*Percent Proportion*), C. Rasio Variasi (*Variance Ration*), dan 2. Tipe Data Interval dan Rasion; A. Rentang (*Range*), B. Variasi (*Variance*), C. Deviasi Baku (*Standard Deviation*).



```python
#Categorical Proportion
print(raw_data['Produk'].value_counts()/raw_data.shape[0])
```

```python
#Range
print(raw_data['Pendapatan'].max() - raw_data['Pendapatan'].min())
```

```python
#Variance
print(raw_data['Pendapatan'].var()) #using pandas but it counts the sample
print(np.var(raw_data['Pendapatan'])) #using numpy but it counts the population
print(raw_data['Pendapatan'].var(ddof=0)) #using ddof to count the population
```

```python
print(raw_data['Pendapatan'].std()) 
print(np.std(raw_data['Pendapatan'], ddof = 1))
```

Beberapa asumsi yang harus dipenuhi untuk menggunakan **Korelasi Pearson** diantaranya adalah: 1. Nilai berskala interval/rasio, 2. Terdapat hubungan linier antara kedua variabel, yaitu kita dapat menggambarkan hubungan kedua variabel sebagai garis lurus, 3. Kedua variabel berdistribusi normal, 4. Homoskedastis, atau data berdistribusi seragam dalam garis regresi

1    = Perfect Positive Correlation,
0.9  = HIght Positive Correlation,
0.5  = Low Positive Correlation,
0    = No Correlation,
-0.5 = Low Negative Correlation,
-0.9 = High Negative Correlation,
-1   = Perfect Negative Correlation. 

```python
print(raw_data.corr())
```

**Korelasi Spearman** atau sering juga disebut sebagai Spearman's rank correlation adalah pengukuran korelasi non-parametrik. Artinya kita mencoba mengukur hubungan antara kedua variabel tanpa menghiraukan asumsi seperti distribusi dari kedua variabel dan asumsi lainnya. Secara kriteria memiliki kemiripan dengan korelasi Pearson walau korelasi Spearman bisa digunakan untuk data bertipe ordinal.

```python
print(raw_data.corr(method='spearman'))
```

**Korelasi Kendall** atau sering juga disebut juga sebagai Kendall's tank correlation atau korelasi Tau (τ) adalah pengukuran korelasi non-parametrik.

```python
print(raw_data.corr(method='kendall'))
```

**Regresi linier** sederhana atau disebut juga sebagai regresi linier bivariat adalah model statistika yang digunakan untuk melakukan estimasi nilai variabel dependen/variabel tak bebas dengan menggunakan satu variabel independen/variabel bebas. Dengan merumuskan bahwa satu variabel memiliki pengaruh terhadap variabel lain, kita dapat mengukur seberapa besar dan signifikan variabel bebas dalam menentukan nilai variabel tak bebas.

```python
import statsmodels.api as sm
nilai_y = raw_data[['Total']]
nilai_x = sm.add_constant(raw_data['Pendapatan'])
model_regresi = sm.OLS(endog=nilai_y, exog=nilai_y).fit()
model_regresi.summary()
```

```python
def is_prime(number):
    if number <= 1:
        return False
    
    for factor in range(2, number):
        if number % factor == 0:
            return False

    return True

def print_primes(n):
    for number in range(1, n):
        if is_prime(number):
            print('%d is prime' % number)
```

```python
print_primes(42)
```
