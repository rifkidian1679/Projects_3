```python
data = pd.read_csv('1.02. Multiple linear regression.csv')
data.head()
```

```python
data.describe()
```

```python
x = data[['SAT', 'Rand 1,2,3']]
y = data['GPA']
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
reg.score(x,y)
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

```python
def adjusted_r2(x,y):
  r2 = reg.score(x,y)
  n = x.shape[0]
  p = x.shape[1]
  print(1-(1-r2)*(n-1)/(n-p-1))
```

```python
adjusted_r2(data[['SAT', 'Rand 1,2,3']], data['GPA'])
```

```python
from sklearn.feature_selection import f_regression

f_stats, p_val = f_regression(x,y)
print('Given 2 features above, we get {0} for F-Statistics,\nand {1} for the p-value'.format(f_stats.round(3), p_val.round(3)))
'''
with univariate p-values 0.05, it means that every features that has p-val of more than 0.05
should be deleted because it has no significant power to predict the target 
'''
```

```python
reg_summary = pd.DataFrame(data=x.columns.values, columns=['Features'])
reg_summary['Coefficient'] = reg.coef_
reg_summary['P-Value'] = p_val.round(3)
reg_summary
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
x_scaled = scaler.transform(x)
x_scaled
```

```python
print(reg.fit(x_scaled,y))
print(reg.coef_)
print(reg.intercept_)
```

```python
reg_summary = pd.DataFrame(data=[['SAT'], ['Rand 1,2,3'], ['Bias']], columns=['Features'])
reg_summary['Weights'] = reg.coef_[0], reg.coef_[1], reg.intercept_
reg_summary
```

```python
new_data = pd.DataFrame(data=[[1759,2], [1653,3]],columns=['SAT', 'Rand 1,2,3'])
new_data_scaled = scaler.transform(new_data)
a, b = reg.predict(new_data_scaled)
print('Both students are predicted to have a GPA of {0} and {1}'.format(a.round(2),b.round(2)))
```

```python
#what if we use only the SAT score to predict the GPA
reg_simple = LinearRegression()
x_simple_matrix = x_scaled[:,0].reshape(-1,1)
reg_simple.fit(x_simple_matrix,y)
c, d = reg_simple.predict(new_data_scaled[:,0].reshape(-1,1))
print('Both students are predicted to have a GPA of {0} and {1}'.format(c.round(2), d.round(2)))
```

**Create Dummy Variables**

```python
cleaned_data.head()
```

```python
#cleaned_data = cleaned_data.drop(['Model'], axis=1)
cleaned_data = pd.get_dummies(cleaned_data, drop_first=True)
cleaned_data.head()
```

```python
cleaned_data.columns
```

```python
cols = ['Log Price', 'Mileage', 'EngineV', 'Registration', 'Year', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol']

preprocessed_data = cleaned_data[cols]
preprocessed_data.head()
```

**Create Linear Regression Model**

```python
targets = preprocessed_data['Log Price']
inputs = preprocessed_data.drop(['Log Price'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
plt.scatter(y_train, y_hat)
plt.title('Linreg Prediction', fontsize=20)
plt.xlabel('Targets (y_train)', fontsize=14)
plt.ylabel('Predictions (y_hat)', fontsize=14)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF', fontsize=20)
```

```python
print(reg.score(x_train, y_train))
print(reg.coef_)
print(reg.intercept_)
```

```python
reg_summary = pd.DataFrame(data=inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
yhat_test = reg.predict(x_test)
plt.scatter(y_test, yhat_test, alpha=0.2)
plt.title('Linreg Prediction', fontsize=20)
plt.xlabel('Targets (y_test)', fontsize=14)
plt.ylabel('Predictions (yhat_test)', fontsize=14)
plt.show()
```

```python
data_performance = pd.DataFrame(np.exp(yhat_test), columns=['Predictions'])
data_performance.head()
```

```python
y_test = y_test.reset_index(drop=True)
data_performance['Targets'] = np.exp(y_test)
data_performance.head()
```

```python
data_performance['Residuals'] = data_performance['Targets'] - data_performance['Predictions']
data_performance.head()
```

```python
data_performance['Difference (%)'] = np.absolute(data_performance['Residuals']/data_performance['Targets']*100)
data_performance.head()
```
